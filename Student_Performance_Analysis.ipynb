```python
# Student Performance Data Analysis

# Objective:
# Analyze student marks dataset and find the relationship between study hours and performance.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Name": [
        "Rahul",
        "Ananya",
        "Kiran",
        "Priya",
        "Arjun",
        "Meera",
        "Vikram",
        "Sneha"
    ],

    "Gender": [
        "Male",
        "Female",
        "Male",
        "Female",
        "Male",
        "Female",
        "Male",
        "Female"
    ],

    "Math": [
        85,92,65,78,
        55,95,72,88
    ],

    "Science": [
        90,88,70,85,
        60,96,75,90
    ],

    "English": [
        78,95,60,80,
        58,92,70,85
    ],

    "Study_Hours": [
        5,6,3,4,
        2,7,4,6
    ]
}


df = pd.DataFrame(data)


df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Average"] = (
    df["Math"]
    +
    df["Science"]
    +
    df["English"]
) / 3


df
```

```python
plt.figure(
    figsize=(10,5)
)


sns.barplot(
    x="Name",
    y="Average",
    data=df
)


plt.title(
    "Average Marks of Students"
)


plt.xlabel(
    "Student Name"
)


plt.ylabel(
    "Average Marks"
)


plt.xticks(
    rotation=45
)


plt.show()
```

```python
plt.figure(
    figsize=(8,5)
)


sns.scatterplot(
    x="Study_Hours",
    y="Average",
    data=df,
    s=100
)


plt.title(
    "Relationship Between Study Hours and Marks"
)


plt.xlabel(
    "Study Hours"
)


plt.ylabel(
    "Average Score"
)


plt.show()
```

```python
subjects = [
    "Math",
    "Science",
    "English"
]


average_subject_marks = (
    df[subjects].mean()
)


average_subject_marks.plot(
    kind="bar"
)


plt.title(
    "Average Marks By Subject"
)


plt.xlabel(
    "Subjects"
)


plt.ylabel(
    "Marks"
)


plt.show()
```
